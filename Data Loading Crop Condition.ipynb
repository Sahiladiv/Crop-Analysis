```python
import numpy
import pickle
import cv2
import os
import matplotlib.pyplot
import random
from sklearn.model_selection import train_test_split
import keras
from tensorflow.keras.layers import Dense, Dropout, Activation, Flatten, Conv2D, MaxPooling2D
```

```python
folder_path = "Images/Condition"
CATEGORIES = os.listdir(folder_path)
CATEGORIES
```

```python
DIR = "Images/Condition"
```

```python
training_data = []
def create_training_data():
    
    for category in CATEGORIES:
        path = os.path.join(DIR, category)
        class_number = CATEGORIES.index(category)
        try:
            for imgs in os.listdir(path):
                img_array = cv2.imread(os.path.join(path,imgs))
                dim = (50,50)
                img_array = cv2.resize(img_array, dim)
                training_data.append([img_array, class_number])
        except Exception as ex:
            pass
```

```python
l1 = create_training_data()
```

```python
random.shuffle(training_data)
```

```python
l = []
f = []

for features, label in training_data:
    l.append(label)
    f.append(features)

f = numpy.array(f).reshape(-1,50,50,3)
l = numpy.array(l)
```

```python
pickle_out = open("condition_features_of_plant.pickle","wb")
pickle.dump(f,pickle_out)
pickle_out.close()
```

```python
pickle_out = open("condition_labels_of_plants.pickle","wb")
pickle.dump(l,pickle_out)
pickle_out.close()
```
